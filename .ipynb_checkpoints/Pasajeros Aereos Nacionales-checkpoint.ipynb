# Viajeros aereos nacionales colombia
Para este trabajo se usó el dataset de [viajeros aereos nacionales en colombia](https://www.datos.gov.co/Comercio-Industria-y-Turismo/Pasajeros-Aereos-Nacionales/idmw-pexi) disponible en la plataforma de [datos abirtos](https://www.datos.gov.co) del gobierno de colombia.

## La pregunta o problema
¿Podemos predecir cuantos viajeros aereos nacionales llegarán a un municipió determinado del pais?

## Aclaraciones iniciales
Dada la naturaleza de la información disponible para cada municipio se concluye que necesitamos aplicar tecnicas de prediccón de series de tiempo, sin embargo, para este trabajo no se usaran tecnicas relacionadas al analisis de series de tiempo, en su lugar, se usaran regresiones lineales unicamente con el fin de hallar tendencias en el numero de viajeros de cada municipio.

Tambien cabe aclarar que para el concurso **datos a la u** si es indispensable aplicar tecnicas de predicción de series de tiempo.

```python
# Formatear fechas
data_clean['FECHA'] = data_clean.apply(lambda fila: datetime(fila['AÑO'], fila['MES'], 1), axis=1)
# Eliminar columna año y columna mes y las colomnuas de latitud y longitud
data_clean = data_clean.drop(['AÑO', 'MES', 'LATITUD', 'LONGITUD'], axis=1)
data_clean = data_clean[['FECHA', 'NOMBRE-DEPARTAMENTO', 'MUNICIPIO', 'PASAJEROS']]
# Renombra nombre de columnas
data_clean.columns = ['FECHA', 'DEPARTAMENTO', 'MUNICIPIO', 'PASAJEROS']
data_clean.head(5)
```

```python
# Cambia el nombre de bogota a uno mas sencillo
data_clean['DEPARTAMENTO'] = data_clean['DEPARTAMENTO'].replace('BOGOTÁ, D. C.', 'BOGOTA')
data_clean['MUNICIPIO'] = data_clean['MUNICIPIO'].replace('BOGOTÁ, D.C.', 'BOGOTA')

# Cambia el nombre de san andres a uno mas sencillo
data_clean['DEPARTAMENTO'] = data_clean['DEPARTAMENTO'].replace('ARCHIPIÉLAGO DE SAN ANDRÉS, PROVIDENCIA Y', 'SAN ANDRES Y PROVIDENCIA')
```

```python
# Creamos un diccionario para acceder mas facil a la info de los municipios sin errores de digitacion
data_clean['MUNICIPIO'].unique()
```

```python
# Funcion para eliminar tildes
def elimina_tildes(cadena):
    s = ''.join((c for c in unicodedata.normalize('NFD',cadena) if unicodedata.category(c) != 'Mn'))
    s = s.lower()    
    s = s.replace(' ', '_')
    return s

# Quita los espacios y tildes del nombre de cada municipio    
data_clean['MUNICIPIO'] = [elimina_tildes(dep) for dep in data_clean['MUNICIPIO']]
data_clean['MUNICIPIO']
```

```python
# Campos del enum
enumfields = " ".join(mun for mun in data_clean['MUNICIPIO'].unique())
enumfields
```

```python
# Crea un enum con los datos de todos los municipios disponibles
municipios = Enum('municipios', enumfields)
```

```python
# Quita los espacios y tildes del nombre de cada departamento    
data_clean['DEPARTAMENTO'] = [elimina_tildes(dep) for dep in data_clean['DEPARTAMENTO']]
data_clean['DEPARTAMENTO']
```

```python
# Campos del enum
enumfields = " ".join(mun for mun in data_clean['DEPARTAMENTO'].unique())
enumfields
```

```python
# Crea un enum de todos los departamentos con datos disponibles
departamentos = Enum('departamentos', enumfields)
```

```python
# Funcion para obtener los datos de un municipio determinado
def getMunicipioInfo(municipio, departamento=None):
    dep = 0
    if (departamento == None):
        dep = data_clean
    else:
        dep = data_clean.loc[data_clean['DEPARTAMENTO'] == departamento.name]
    mun = dep.loc[dep['MUNICIPIO'] == municipio.name]
    return mun

# Usamos los enums que definimos anteriormente para facilitar la vida al usuario y disminuir las probabilidades de que escriba mal el nombre de un municipio o departamento
getMunicipioInfo(municipios.barrancabermeja, departamentos.santander).head(5)
```

```python
# Obtiene la serie de tiempo para un municipio determinado
def getMunicipioTimeSerie(municipio, departamento=None, verbose=False):
    mun = getMunicipioInfo(municipio, departamento)
    '''
    Lo bueno de usar como indice pd.DatatimeIndex es que podemos filtrar rangos
    de fechas mucho mas facilmente por ejemplo si hacemos:
    mun['2014'] obtenemos unicamente las filas del 2014
    
    Referencia: https://stackoverflow.com/questions/36506149/create-datetimeindex-in-pandas
    '''
    mun.index = pd.DatetimeIndex(mun.FECHA)
    # Quita la información relacionada al municipio y al departamento de este data frame si el parametro verbose == False
    if (verbose):
        mun = mun.iloc[:, [1, 2, 3]]
    else:
        mun = mun.iloc[:, [3]]
    return mun

getMunicipioTimeSerie(municipios.medellin, verbose=True).head()
```

```python
# Graficamos el numero de pasajeros aereos nacionales de san andres y providencia desde el principio de los tiempos (desde el princpio de los datos disponibles)
plt.plot(getMunicipioTimeSerie(municipios.san_andres, departamentos.san_andres_y_providencia), label=municipios.san_andres.name)
plt.plot(getMunicipioTimeSerie(municipios.providencia, departamentos.san_andres_y_providencia), label=municipios.providencia.name)

plt.legend()
plt.show()
```

```python
# Graficamos el numero de pasajeros aereos nacionales de los distintos municipios de antioquia desde el principio de los tiempos (desde el princpio de los datos disponibles)
plt.plot(getMunicipioTimeSerie(municipios.medellin, departamentos.antioquia), label=municipios.medellin.name)
plt.plot(getMunicipioTimeSerie(municipios.rionegro, departamentos.antioquia), label=municipios.rionegro.name)

plt.legend()
plt.show()
```

```python
plt.plot(getMunicipioTimeSerie(municipios.leticia, departamentos.antioquia), label=municipios.leticia.name)
plt.plot(getMunicipioTimeSerie(municipios.apartado, departamentos.antioquia), label=municipios.apartado.name)
plt.plot(getMunicipioTimeSerie(municipios.caucasia, departamentos.antioquia), label=municipios.caucasia.name)

plt.legend()
plt.show()
```

```python
# Grafica el numero de visitantes aereos locales año por año en medellin
# plt.figure(1) # Esta invocación es opcional debido a que por defecto se llama detras de escenas
numOfYears = (2016 - 2004)+1 # Los datos estan en este rango de años
medTs = getMunicipioTimeSerie(municipios.medellin, departamentos.antioquia)
plt.figure(figsize=(15, 40))
for i in range(numOfYears):
    plt.subplot(numOfYears, 1, i+1)
    year = '20' + ('%02d'% (i+4))
    plt.plot(medTs[year])
    plt.title(year)

plt.legend()
plt.plot()
```

# Aplicando regresión lineal a los datos
Vamos ahora a aplicar una regresión lineal a los datos de rionegro para hallar una tendencia, notese que esto no es con fines predictivos, para ello en su lugar seria aplicar metodos mas aptos para analizar series de tiempo

```python
# Importando las librerias relacionadas a la regresion
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
import pickle # Para serializar y des-serializar objetos
```

```python
regr = linear_model.LinearRegression() # Crear el objeto de regresion lineal
# lin = regr.fit(, medTs.values)
X = np.arange(len(medTs.index)).reshape(-1, 1)
y = medTs.values.reshape(-1, 1) 

lin = regr.fit(X, y)
# Returns the coefficient of determination R^2 of the prediction.
# The coefficients
print('Coefficients: \n', lin.coef_)
print('Intercept: ', lin.intercept_ )
print("R2 score", lin.score(X, y))
```

```python
'''
Graficamos el modelo de esta forma, para evitar tener que acceder al coeficiente e intercepto del modelo y realizar
la graficación manualmente.
'''
lin_y = regr.predict(X)
plt.plot(lin_y)
```

```python
# Calcula y grafica el modelo lineal para cada municipio
models = []

# Crea un lienzo donde dibujar todas las graficas
numOfCols = 2
numOfRows = len(municipios)/numOfCols
graphWidth = 7.5
graphHeight = 5
plt.figure(figsize=(graphWidth*numOfCols, graphHeight*numOfRows))

# Calcula y grafica los modelos lineales
i = 1 # Index used by the for loop
for mun in municipios:
    # Obtiene la timeserie de cada municipio
    ts = getMunicipioTimeSerie(mun)
    # Los ajusta para poder ser usados por el metodo fit
    X = np.arange(len(ts.index)).reshape(-1, 1)
    y = ts.values.reshape(-1, 1)
    
    # Calcula la regresion lineal
    lin = regr.fit(X, y)
    # Agrega el modelo entrenado a un array de modelos para guardarlo persistentemente usando pickle
    models.append(lin)
    
    # Graficamos la regresion lineal calculada
    '''
    Para graficar la linea del modelo lineal calculada, en lugar de usar los coeficientes
    del modelo entrenado, se decidió usar el metodo predict() solo que en lugar
    de pasarle datos nuevos, le pasamos los mismo datos que usamos para entrenar a
    el modelo, de esta forma la grafica es mas sencilla.
    '''
    lin_y = regr.predict(X)
    # Crea una sub grafica para cada municipio
    plt.subplot(numOfRows, numOfCols, i)    
    plt.plot(lin_y, label='Regresion lineal')
    plt.plot(ts.values, label='Pasajeros aereos nacionales')
    plt.legend()
    plt.title(mun.name)
    i = i + 1
    

plt.plot()
```

```python
# Usaremos pickle para serializar y des-serializar el array de modelos que entrenamos
# Pickle example: https://docs.python.org/3/library/pickle.html#examples

with open('models.pickle', 'wb') as f:
    # Pickle (Serializes) the trained models by using the highest protocol available.
    pickle.dump(models, f, pickle.HIGHEST_PROTOCOL)
```
